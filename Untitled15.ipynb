## Divide Dataset into Train and Test Sets (Random Sampling)

### Subtask:
Prepare the existing `strat_train_set` and `strat_test_set` by handling missing values, encoding categorical features, and splitting them into feature (X) and target (y) sets.


**Reasoning**:
The first step is to impute missing values in the 'total_bedrooms' column for both the training and test sets. According to the instructions, the mode should be calculated from the `strat_train_set` only and then applied to both sets to prevent data leakage.



```python
mode_total_bedrooms = strat_train_set['total_bedrooms'].mode()[0]

strat_train_set['total_bedrooms'].fillna(mode_total_bedrooms, inplace=True)
strat_test_set['total_bedrooms'].fillna(mode_total_bedrooms, inplace=True)

print(f"Mode of total_bedrooms from strat_train_set: {mode_total_bedrooms}")
print("Missing values in strat_train_set after imputation:")
print(strat_train_set['total_bedrooms'].isnull().sum())
print("Missing values in strat_test_set after imputation:")
print(strat_test_set['total_bedrooms'].isnull().sum())
```

**Reasoning**:
The previous step successfully imputed missing values but generated a `FutureWarning` due to `inplace=True` on a chained assignment. This step addresses the warning by re-assigning the filled column and then proceeds to one-hot encode the `ocean_proximity` categorical feature, convert the encoded results to DataFrames, drop original columns, concatenate features, and finally split the data into feature (X) and target (y) sets for both training and testing, as per the subtask instructions.



```python
from sklearn.preprocessing import OneHotEncoder

# Impute missing values (addressing FutureWarning)
mode_total_bedrooms = strat_train_set['total_bedrooms'].mode()[0]
strat_train_set['total_bedrooms'] = strat_train_set['total_bedrooms'].fillna(mode_total_bedrooms)
strat_test_set['total_bedrooms'] = strat_test_set['total_bedrooms'].fillna(mode_total_bedrooms)

# One-hot encode 'ocean_proximity'
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit on training data and transform both train and test data
ocean_proximity_encoded_train = encoder.fit_transform(strat_train_set[['ocean_proximity']])
ocean_proximity_encoded_test = encoder.transform(strat_test_set[['ocean_proximity']])

# Create DataFrames from encoded features
ocean_proximity_cols = encoder.get_feature_names_out(['ocean_proximity'])
df_ocean_proximity_train = pd.DataFrame(ocean_proximity_encoded_train, columns=ocean_proximity_cols, index=strat_train_set.index)
df_ocean_proximity_test = pd.DataFrame(ocean_proximity_encoded_test, columns=ocean_proximity_cols, index=strat_test_set.index)

# Prepare numerical features by dropping original 'ocean_proximity' and target column
X_train_numerical = strat_train_set.drop(['ocean_proximity', 'median_house_value'], axis=1)
X_test_numerical = strat_test_set.drop(['ocean_proximity', 'median_house_value'], axis=1)

# Concatenate numerical and one-hot encoded features
X_train = pd.concat([X_train_numerical, df_ocean_proximity_train], axis=1)
X_test = pd.concat([X_test_numerical, df_ocean_proximity_test], axis=1)

# Define target variables
y_train = strat_train_set['median_house_value']
y_test = strat_test_set['median_house_value']

print("Preprocessing complete. X_train, y_train, X_test, y_test are prepared.")
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

## Apply Standard Scaling

### Subtask:
Apply Standard Scaling to all numerical features in both the training and testing sets.


**Reasoning**:
To apply Standard Scaling, I will import `StandardScaler`, initialize it, fit it on `X_train` to learn scaling parameters, and then transform both `X_train` and `X_test` using the fitted scaler, storing the results in new dataframes.



```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the training features and transform X_train
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)

# Transform the testing features using the fitted scaler
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

print("Standard Scaling applied to X_train and X_test.")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_test_scaled shape: {X_test_scaled.shape}")
print("First 5 rows of X_train_scaled:")
print(X_train_scaled.head())
```

**Reasoning**:
With the numerical features now scaled, the next logical step is to train a Linear Regression model using the preprocessed training data (`X_train_scaled` and `y_train`). After training, the model's performance will be evaluated on the preprocessed test data (`X_test_scaled` and `y_test`) by calculating the Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), and R2 Score.



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Initialize the Linear Regression model
lin_reg = LinearRegression()

# Train the model using the scaled training data
lin_reg.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred = lin_reg.predict(X_test_scaled)

# Calculate evaluation metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Linear Regression Model Performance on Test Set:")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R2 Score: {r2:.2f}")
```
